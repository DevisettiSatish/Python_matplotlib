```python
import matplotlib.pyplot as plt
```

```python
# Scatter Plot
x=[1,4,8,10,15]
y=[3,5,7,9,13]

plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(5,4))
x=[1,4,8,10,15]
y=[3,5,7,9,13]


plt.title('scatter plot')
plt.xlabel('x_label')
plt.ylabel('y_label')

plt.xticks([])
plt.yticks([])


plt.scatter(x , y , color = 'r' , marker = 'd')

plt.show()
```

```python
# Example:-2

plt.figure(figsize=((4,5)))

a=[10,40,60,100]
b=[12,25,35,78]


plt.title('scatter plot_1')
plt.xlabel('x_label')
plt.ylabel('y_label')

plt.xticks([])
plt.yticks([])


plt.scatter(a,b,color='b',marker='d')
plt.show()
```

```python
# line plot

x=[1,4,8,10,14]
y=[3,5,7,11,9]

plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=((5,3)))
x=[1,4,8,10,14]
y=[3,5,7,11,9]


plt.title('line plot')
plt.xlabel('x_label')
plt.ylabel('y_label')

plt.plot(x,y,color='g',marker='s')

plt.show()
```

```python
# Legend Functions

days=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]

plt.figure(figsize = (5,2))
plt.plot(days,a)
plt.plot(days,b)
plt.plot(days,c)

plt.show()
```

```python
# bar Chart

mobile=['realme','vivo','Apple','samsunbg']
profits=[25000,23000,30000,24000]
plt.bar(mobile,profits ,color='g')
plt.show()
```

```python
languages=['python','sql','powerbi','tableau']
profit=[23000,26000,27000,30000]

plt.bar(languages,profit,color='r')
plt.show()
```

```python
# pie char
profits = [60 , 40 , 30 , 10 , 1]

names = ['Avenger' , 'Avatar' , 'SR' , 'abc' , 'shark']

plt.pie(profits , labels = names , shadow = True , autopct = '%1.1f%%')

plt.show()
```

```python
# subplots
plt.figure(figsize = (8,3))

days=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]


# plot days and a:

plt.subplot(1,3,1)
plt.title('days_a')
plt.plot(days,a)


plt.subplot(1,3,2)
plt.title('days_b')
plt.plot(days,b)


plt.subplot(1,3,3)
plt.title('days_c')
plt.plot(days,c)

plt.show()
```
